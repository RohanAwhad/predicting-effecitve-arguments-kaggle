```python
import pandas as pd

COMP_DIR = '../data/'
df = pd.read_parquet(COMP_DIR + 'train_UPD.parquet')
```

```python
df.head()
```

```python
from transformers import GPT2Tokenizer, GPT2ForSequenceClassification
```

```python
model_path = 'gpt2'
# model_path = '../model/sample_gpt2'

tokenizer = GPT2Tokenizer.from_pretrained(model_path)
model = GPT2ForSequenceClassification.from_pretrained(model_path, num_labels=3)
```

```python
# from transformers import AutoModelForSequenceClassification, AutoTokenizer

# model_path = "../model/sample_distilbert_base"

# tokenizer = AutoTokenizer.from_pretrained(model_path)
# model = AutoModelForSequenceClassification.from_pretrained(model_path, num_labels=3)
```

```python
id2label = {
    0: 'Ineffective',
    1: 'Adequate',
    2: 'Effective',
}
label2id = dict((v, k) for k, v in id2label.items())

label2id
```

```python
df['labels'] = df['discourse_effectiveness'].apply(lambda x: label2id[x])
df.head()
```

### Random Inference

```python
model.config.id2label = id2label
model.config.label2id = label2id
```

```python
example = df.loc[15542, 'discourse_text_UPD']
example
```

```python
example = ' '.join(example.split())
```

```python
df.loc[15542, 'discourse_effectiveness']
```

```python
inputs = tokenizer(example, return_tensors='pt')
```

```python
tokenizer.tokenize(example)
```

```python
logits = model(**inputs).logits.detach().cpu().numpy()
logits
```

```python
id2label[logits.argmax(axis=1)[0]]
```

```python
from transformers import pipeline

nlp = pipeline('text-classification', model=model, tokenizer=tokenizer)
```

```python
nlp(example)
```
